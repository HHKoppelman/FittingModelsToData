Fitting a straight line:   
    $y = mx + b$

```python
def cov_ellipse(x,y,sx,sy,rxy):
    """
    x,y = center of ellipse
    sx, sy = standard deviation in x and y
    rxy = correlation between sx and sy; such that cov(x,y) = rxy * sx * sy
    
    
    Source:
    https://cookierobotics.com/007/
    """
    from math import sqrt, atan2, degrees
    a,b,c = sx**2,sx*sy*rxy,sy**2
    
    
    l1 = sqrt((a+c)/2 + sqrt((a/2-c/2)**2 + b**2))
    l2 = sqrt((a+c)/2 - sqrt((a/2-c/2)**2 + b**2))
    angle = 0 if (b==0 and a>=c) else (90 if b==0 and a<c else degrees(atan2(l1**2-a,b)))
    return Ellipse(xy=(x,y),
                width=l1*2, height=l2*2,
                angle=angle, edgecolor='black')
```

```python
def Log_Likelihood(params, data):
    b,m = params
    
    log_likelihood = 0

    for x,y,sx,sy,rxy in data:
    
        Z_i = np.array([x,y])
        v_hat = 1/np.sqrt(1+m**2)*np.array([-m,1])
        theta = np.arctan(m)
        Delta_i = (v_hat.T).dot(Z_i)  - b * np.cos(theta)# Transpose is redundant..

        S_i = np.array([[sx**2,rxy*sx*sy],[rxy*sx*sy,sy**2]])
        Sigma_i_squared = (v_hat.T).dot(S_i.dot(v_hat))
        
        log_likelihood -= 0.5* Delta_i**2 / Sigma_i_squared
    return -log_likelihood
```

```python
params0 = [600,20]
data = np.array([x,y,sx,sy,rxy]).T
res = minimize(Log_Likelihood, params0, args=(data))
b,m = res.x
print(b,m)
```

We will now fit the following mixture model:

$$\mathcal{L} = \prod_i^N \bigg[ (1-P_b)p_{fg} + P_bp_{bg} \bigg],$$
where  
$$p_{fg} = \frac{1}{\sqrt{\det{S_i}}} \exp\bigg(-\frac{\Delta_i}{2\Sigma_i}\bigg)$$
(ignoring a factor of $1/2\pi$, see notes of the 'Chapter' for more information).  
And  

\begin{equation}
p_{bg} = \frac{1}{\sqrt{\det{V_b}}} \exp\bigg(-\frac{1}{2}  
\bigg[Z_i - Y_b\bigg]^T
V_b^{-1}
\bigg[Z_i - Y_b\bigg]
\bigg),
\end{equation}
where 
$$Y_b = [Y_1, Y_2]$$
is an arbitrary central point of the distribution, and 

\begin{equation}
V_b = 
\begin{bmatrix}
\sigma_x^2+V^2 & 0 \\
0 & \sigma_y^2+V^2 
\end{bmatrix},
\end{equation}
an arbitrary large covariance matrix (with only diagonal elements)

```python
def p_foreground(xyz, b, m):
        x,y,sx,sy,rxy = xyz
    
        Z_i = np.array([x,y])
        v_hat = 1/np.sqrt(1+m**2)*np.array([-m,1])
        theta = np.arctan(m)
        Delta_i = (v_hat.T).dot(Z_i)  - b * np.cos(theta)# Transpose is redundant..

        S_i = np.array([[sx**2,rxy*sx*sy],[rxy*sx*sy,sy**2]])
        Sigma_i_squared = (v_hat.T).dot(S_i.dot(v_hat))
        
        return 1/np.sqrt(np.linalg.det(S_i)) * np.exp(-0.5*Delta_i**2 / Sigma_i_squared)
        
def p_background(xyz, Yb, V):

        x,y,sx,sy,rxy = xyz
    
        Z_i = np.array([x,y])
        Vb = np.array([[sx**2+V**2,0],[0,sy**2+V**2]])
        iVb = np.linalg.inv(Vb)
        
        Delta_i = Z_i - Yb
        
        exp_part = (Delta_i.T).dot(iVb.dot(Delta_i))
        
        return 1/np.sqrt(np.linalg.det(Vb)) * np.exp(-0.5*exp_part)
        

def Log_Likelihood_mixture(params, data):
    b, m, V, Y1, Y2, Pb = params
    
    Yb = np.array([Y1, Y2])

    
    log_likelihood = 0

    for xyz in data:
        
        
        pfg = p_foreground(xyz,b,m)
        pbg = p_background(xyz,Yb,V)

        log_likelihood += np.log( (1-Pb)*pfg + Pb*pbg)

    return log_likelihood
```

```python
def log_prior(params):
    b, m, V, Y1, Y2, Pb = params

    _mr = [0,5]
    _br = [-50,250]
    _Pbr = [0,1]
    _Y1r = [-50,500]
    _Y2r = [-50,700]    
    _Vr = [0,500]
    
    mb_if = (_mr[0] < m < _mr[1]) and (_br[0] < b < _br[1])
    Nuisance_if1 = (_Pbr[0] < Pb < _Pbr[1])and (_Y2r[0] < Y2 < _Y2r[1]) 
    Nuisance_if2 = (_Y1r[0] < Y1 < _Y1r[1]) and (_Vr[0] < V < _Vr[1])
    
    _rlist = [_mr, _br, _Pbr, _Y1r, _Y2r, _Vr]
    norm = sum([np.abs(_r).sum() for _r in _rlist])
    
    if mb_if and Nuisance_if1 and Nuisance_if2:
        return 1/norm
    return -np.inf
```

```python
# def log_probability(x,y,sy,theta):
def log_probability(params, x,y,sx,sy,rxy):  
    """
    params = [b, m, V, Y1, Y2, Pb]
    where m,b are the model parameters, and 
    V, Y1, Y2, Pb, are nuisance parameters that we will marginalize over.
    """
    data = np.array([x,y,sx,sy,rxy]).T
    lp = log_prior(params)
    if ~np.isfinite(lp):
        return -np.inf
    
    return lp + Log_Likelihood_mixture(params, data)
    
```

```python
from multiprocessing import Pool
import emcee
import corner
```

```python
initial_pos = np.array([30, 2, 50,200,200,0.1])*(1+0.2*np.random.normal(size=(20,6)))

nwalkers, ndim = initial_pos.shape


nwalkers
```

```python
with Pool(processes=8) as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(x,y,sx,sy,rxy),pool=pool)
    
    sampler.run_mcmc(initial_pos, 5000, progress=True);
```

```python
fig, axes = plt.subplots(ndim, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
labels = ['b','m','V','Y1','Y2','Pb']
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```
